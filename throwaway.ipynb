```python
im_path = "//data2/saikiran.tedla/hdrvideo/diff/evaluations/stuttgart/under/bistro_01/frame_0000.png"

import cv2
import numpy as np
import os
os.environ["OPENCV_IO_ENABLE_OPENEXR"] = "1"
from PIL import Image
img = np.array(Image.open(im_path))
import matplotlib.pyplot as plt
rescale = np.clip(img*10, 0, 255).astype(np.uint8)
plt.imshow(rescale)
plt.show()
plt.imsave("rescaled.png", rescale)

#show  amask of any pixels that are clipped
clipped0 = np.any(img == 0, axis=-1)   # H x W boolean
print("Number of pixels clipped to 0 in any channel:", np.sum(clipped0))
print("Percentage of pixels clipped to 0 in any channel:", np.mean(clipped0)*100, "%")
plt.imshow(clipped0, cmap="gray")
plt.title("Pixels clipped to 0 in any channel")
plt.show()


#load 
hdr_img = "/data2/saikiran.tedla/hdrvideo/diff/evaluations/ours_stuttgart/under/bistro_03/frame_0000.exr"
hdr = cv2.imread(hdr_img, cv2.IMREAD_UNCHANGED)[:,:,::-1]
scaled_hdr = np.clip(hdr*50*255, 0, 255).astype(np.uint8)
plt.imshow(scaled_hdr)
plt.show()
plt.imsave("scaled_hdr.png", scaled_hdr)
```
